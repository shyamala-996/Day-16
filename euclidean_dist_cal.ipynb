Euclidean Distance Calculator

```python
import numpy as np
v1 = np.array([2.0, 4.0, 6.0])
v2 = np.array([2.0, 3.0, 4.0])

def euclidean_distance(v1: np.ndarray, v2: np.ndarray) -> float:
    diff = v1 - v2
    sq   = diff ** 2
    sumsq = np.sum(sq)
    dist = np.sqrt(sumsq)
    return dist

print("Distance:", euclidean_distance(v1, v2))
```
