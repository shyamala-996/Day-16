From Data to Matrix

Create Raw Data:

```python
import numpy as np

Student_grades = [
    ['Shyam', 97, 95],
    ['Ram', 91, 92],
    ['Sita', 85, 87]    
    ]

names = [row[0] for row in Student_grades]
scores = [row [1:] for row in Student_grades]
```

Convert to Matrix:

```python
score_matrix = np.array(scores)
print("Original score matrix:")
print(score_matrix)
```

Add a new feature:

```python
sports_score = np.array([88, 92, 79]).reshape(-1, 1)

Overall_matrix = np.hstack((score_matrix, sports_score))
print("\nAfter adding sports_score column:")
print(Overall_matrix)
```

Calculate a new feature:

```python
average_score = Overall_matrix.mean(axis=1).reshape(-1, 1)
avg_matrix = np.hstack((Overall_matrix, average_score))
print("\nAfter adding average_score column:\n", avg_matrix)
```

Scale a feature:

```python
scaled_matrix = avg_matrix.copy()
scaled_matrix[:,0] = scaled_matrix[:,0] * 1.1
print("\nAfter scaling math_score by factor 1.1:")
print(scaled_matrix)
```

```python
print("\nNames with final data:")
for name, row in zip(names, scaled_matrix):
    print(f"{name}: {row}")
```
