Basic Data Scaling

```python
import numpy as np
M = np.array([
    [10, 200, 3000],
    [12, 190, 4000],
    [11, 210, 3500]
])
print("M:\n", M)
```

Mean & Standard Deviation

```python
mean = np.mean(M, axis=0)
std_dev = np.std(M, axis=0)
print("\nMean:\n", mean)
print("\nStandard Deviation:\n", std_dev)
```

Scale: subtract mean, then divide by std

```python
M_scaled = (M - mean) / std_dev
print("\nScaled matrix (each feature now has mean 0 and std 1):\n", M_scaled)
```

```python
print("\nCheck means after scaling (should be ~0):", M_scaled.mean(axis=0))
print("Check stds after scaling (should be ~1):", M_scaled.std(axis=0))
```
