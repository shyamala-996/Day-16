Simple Neural Network Layer Simulation

```python
import numpy as np
input_vector = np.array([2, 3, 4])

weights_matrix = np.random.rand(3, 2)  # 3 inputs â†’ 2 outputs
bias_vector = np.array([0.1, 0.2], dtype=float)  # one bias per output unit

print("Input vector:", input_vector)
print("\nWeights matrix:\n", weights_matrix)
print("\nBias vector:", bias_vector)

```

Perform Linear Transformation:

```python
output = input_vector @ weights_matrix + bias_vector
print("\nOutput vector:", output)   
```

Add Activation Function:

```python
def sigmoid(x):
    return 1.0 / (1.0 + np.exp(-x))
```

```python
activated_output = sigmoid(output)
print("\nAfter applying sigmoid activation:", activated_output)

# final result
print("\nFinal output vector:", activated_output)
```
